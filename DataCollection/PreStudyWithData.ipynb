```python
import glob
import itertools
import json
import os
import pandas as pd
import numpy as np
import scipy
from scipy import stats
```

```python
preStudyDf = pd.read_csv("PreStudy.csv")
preStudyDf = preStudyDf.sort_values(by=["What is your Participant Id?", "Timestamp"])
preStudyDf

preStudyDf = preStudyDf.drop_duplicates(subset=["What is your Participant Id?"], keep="first")
dropbyUserId = preStudyDf.loc[preStudyDf["What is your Participant Id?"] == 74]
preStudyDf = preStudyDf.drop(dropbyUserId.index)
preStudyDf = preStudyDf.rename(columns={"What is your Participant Id?": "userID"})
preStudyDf
```

```python
articleDF = pd.read_csv("ArticleExperimentData.csv")
articleDF = articleDF.sort_values(by="userID")
articleDF = articleDF.drop(columns=['Unnamed: 0'])
articleDF

```

```python
allData = pd.merge(preStudyDf, articleDF, on="userID")

allData.to_csv("PreStudy_with_ArticleData.csv")
```

```python
#change mode from static, responsive to 0 and 1

mask = allData['Mode'] == 'Static'
allData.loc[mask, 'Mode'] = int(0)

mask = allData['Mode'] == 'Responsive'
allData.loc[mask, 'Mode'] = int(1)


mask = allData['How many times have you used an AR headset?'] != '0'
allData.loc[mask, 'How many times have you used an AR headset?'] = 1

mask = allData['How many times have you used an AR headset?'] == '0'
allData.loc[mask, 'How many times have you used an AR headset?'] = 0
allData

mask = allData['Please indicate your proficiency in English:'] == "10 or more years of experience"
allData.loc[mask, 'Please indicate your proficiency in English:'] = 1
mask = allData['Please indicate your proficiency in English:'] == "English is my native language"
allData.loc[mask, 'Please indicate your proficiency in English:'] = 1
mask = allData['Please indicate your proficiency in English:'] != 1
allData.loc[mask, 'Please indicate your proficiency in English:'] = 0

allData.to_csv("PreStudy_with_ArticleDataNew.csv")
```

```python
import statsmodels.api as sm
df = allData

X = df[['Mode', 'How many times have you used an AR headset?', 'Please indicate your proficiency in English:']]

Y = df['Guesses']

X = sm.add_constant(X) # adding a constant

model = sm.OLS(Y.astype(float), X.astype(float)).fit()
predictions = model.predict(X) 

print_model = model.summary()
print(print_model)
```
